### 根據範例檔案的結果：

1. 取出班次一的每一個時間
2. 將班次一的每一個時間用一種資料型態保存
3. 將班次一到五與其所有時間用一種資料型態個別保存

```python
# 1. 取出班次一的每一個時間
import csv


# 開啟 CSV 檔案
with open('./example.csv', newline='',encoding='utf-8') as csvfile:
    # 讀取 CSV 檔案內容
    rows = csv.reader(csvfile)
    headerrow = next(rows)
    first = []
    for row in rows :
        first.append(row[5])

print("班次一 : ",first)
print(type(first))
```

```python
# 2. 將班次一的每一個時間用一種資料型態保存

import csv
from datetime import datetime


# 開啟 CSV 檔案
with open('./example.csv', newline='',encoding='utf-8') as csvfile:
    # 讀取 CSV 檔案內容
    rows = csv.reader(csvfile)
    headerrow = next(rows)
    first = []
    for row in rows :
        firstdate = datetime.strptime(row[5],'%H:%M')
        first.append(firstdate)

print("班次一 : ",first)
```

```python
# 3. 將班次一到五與其所有時間用一種資料型態個別保存
import csv
from datetime import datetime


# 開啟 CSV 檔案
with open('./example.csv', newline='',encoding='utf-8') as csvfile:
    # 讀取 CSV 檔案內容
    rows = csv.reader(csvfile)
    headerrow = next(rows)
    f1,f2,f3,f4,f5,f6,f7,f8,f9,f10,f11 = [],[],[],[],[],[],[],[],[],[],[]
    for row in rows :
        f1_d = datetime.strptime(row[5],'%H:%M')
        f2_d = datetime.strptime(row[6],'%H:%M')
        f3_d = datetime.strptime(row[7],'%H:%M')
        f4_d = datetime.strptime(row[8],'%H:%M')
        f5_d = datetime.strptime(row[9],'%H:%M')
        f6_d = datetime.strptime(row[10],'%H:%M')
        f7_d = datetime.strptime(row[11],'%H:%M')
        f8_d = datetime.strptime(row[12],'%H:%M')
        f9_d = datetime.strptime(row[13],'%H:%M')
        f10_d = datetime.strptime(row[14],'%H:%M')
        f11_d = datetime.strptime(row[15],'%H:%M')
        f1.append(f1_d)
        f2.append(f2_d)
        f3.append(f3_d)
        f4.append(f4_d)
        f5.append(f5_d)
        f6.append(f6_d)
        f7.append(f7_d)
        f8.append(f8_d)
        f9.append(f9_d)
        f10.append(f10_d)
        f11.append(f11_d)
       
print("班次1: ",f1)
print("班次2: ",f2)
print("班次3: ",f3)
print("班次4: ",f4)
print("班次5: ",f5)
print("班次6: ",f6)
print("班次7: ",f7)
print("班次8: ",f8)
print("班次9: ",f9)
print("班次10: ",f10)
print("班次11: ",f11)
```
